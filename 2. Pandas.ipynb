The data set that I am using here contains contains 18625 rows and 11 columns based on _18625_ game reviews obtained from [Kaggle](https://www.kaggle.com/egrinstein/20-years-of-games/downloads/ign.csv/2). The dataset contains the following information about the games.
* __score_phrase__    : how IGN described the game in one word. This is linked to the score it received.
* __title__           : the name of the game.
* __url__             : the URL where you can see the full review.
* __platform__        : the platform the game was reviewed on (PC, PS4, etc).
* __score__           : the score for the game, from 1.0 to 10.0.
* __genre__           : the genre of the game.
* __editors_choice__  : N if the game wasn't an editor's choice, Y if it was. This is tied to score.
* __release_year__    : the year the game was released.
* __release_month__   : the month the game was released.
* __release_day__     : the day the game was released.

# 3.Pandas - Experimenting with dataset[thanksGiving.csv by FiveThirtyEight]

````python
import pandas as pd
thanksGiving = pd.read_csv("thanksGiving.csv")
thanksGiving
````

````python
#finding the unique values from a set of values
thanksGiving["Do you celebrate Thanksgiving?"].unique()
````

````python
thanksGiving["How is the main dish typically cooked?"].unique()
````

````python
#This gives all the columns from the dataframe
thanksGiving.columns[50:]
````

````python
#We check the number of values of each data using the value_counts method and dropna tells whether to display NaN counts
thanksGiving["What is your gender?"].value_counts(dropna = False)
````

````python
#here we assign numeric values to genders female = 1.0 and male = 0.0 and then display the count

import math

def gender_code(gender):
    if isinstance(gender,float) and math.isnan(gender):
        return gender
    return int(gender == "Female")

thanksGiving["gender"] = thanksGiving["What is your gender?"].apply(gender_code)
thanksGiving["gender"].value_counts(dropna = False)
````

````python
thanksGiving["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts(dropna = False)
````

````python
import numpy as np

def cleanIncome(incomeRange):
    if incomeRange == "$200,000 and up":
        return 200000
    elif incomeRange == "Prefer not to answer":
        return np.nan
    elif isinstance(incomeRange, float) and math.isnan(incomeRange):
        return np.nan
    
    incomeRange = incomeRange.replace("," , "").replace("$" , "")
    upperIncome,lowerIncome = incomeRange.split(" to ")
    return (int(upperIncome) + int(lowerIncome))/2

thanksGiving["income"] = thanksGiving["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(cleanIncome)
thanksGiving["income"].value_counts(dropna = False)
````

````python
thanksGiving["What type of cranberry saucedo you typically have?"].value_counts()
````

````python
homemade = thanksGiving[thanksGiving["What type of cranberry saucedo you typically have?"]== "Homemade"]
canned = thanksGiving[thanksGiving["What type of cranberry saucedo you typically have?"]== "Canned"]
````

````python
print("Mean income of people having homemade sauce: $", homemade["income"].mean())
print("Mean income of people having canned sauce: $", canned["income"].mean())
````

````python
#here we make groups based on the type of cranberry sauce people consume
sauceGroups = thanksGiving.groupby("What type of cranberry saucedo you typically have?")
sauceGroups.groups
````

````python
sauceGroups.size()
````

````python
for name, group in sauceGroups:
    print("Group, Group Size: ", name, ",", group.size)
````

````python
#using the aggregate method on the grouped data we can easily find the results
sauceGroups["income"].agg(np.mean)
````

````python
#now we plot the above generated aggregate data
%matplotlib inline
sauceGroups["income"].agg(np.mean).plot(kind = "bar")
````

````python
thanksGiving["What is typically the main dish at your Thanksgiving dinner?"].value_counts()
````

````python
#here we make groups based on multiple attributes or features
multiGroup = thanksGiving.groupby(["What type of cranberry saucedo you typically have?", "What is typically the main dish at your Thanksgiving dinner?"])
multiGroup.groups
````

````python
#here we make a single aggregate over multiple values
multiGroup.agg(np.mean)
````

````python
#now we try and make multiple aggregate over a single parameter income
multiGroup["income"].agg([np.mean, np.median, np.sum, np.std])
````
